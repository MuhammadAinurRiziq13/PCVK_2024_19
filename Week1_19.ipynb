2. Buat garis vertikal dan garis menyilang diagonal pada image keluaran

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

# Membuat garis vertikal di tengah image
for x in range(tinggi):
    image_3[x, int(lebar / 2)] = [255, 255, 255]

# Gabungkan gambar asli dengan gambar yang memiliki garis vertikal
vertical_combined = cv.hconcat((image_2, image_3))

# Membuat tanda silang (garis diagonal)
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
for i in range(tinggi):
    # Garis diagonal dari kiri atas ke kanan bawah
    x = int(i * (lebar / tinggi))
    image_3[i, x] = [255, 255, 255]

    # Garis diagonal dari kanan atas ke kiri bawah
    y = int(lebar - (i * (lebar / tinggi)))
    image_3[i, y - 1] = [255, 255, 255]

# Gabungkan gambar asli dengan gambar yang memiliki tanda silang
cross_combined = cv.hconcat((image_2, image_3))

# Gabungkan kedua hasil di atas secara vertikal
final_output = cv.vconcat((vertical_combined, cross_combined))

# Tampilkan output final
cv2_imshow(final_output)
```

3. Buat garis horisontal berwarna putih dibagian tengah gambar dengan panjang
tertentu

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = image_2.copy()

# Panjang garis yang diinginkan (misalnya, setengah dari lebar gambar)
panjang_garis = lebar // 2

# Posisi awal dan akhir dari garis
start_x = (lebar - panjang_garis) // 2
end_x = start_x + panjang_garis

# Membuat garis horizontal di tengah gambar dengan panjang tertentu
for y in range(start_x, end_x):
    image_3[int(tinggi / 2), y] = [255, 255, 255]

# Gabungkan gambar asli dengan gambar yang memiliki garis horizontal
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

 4.  Buat kotak menggunakan kumpulan pixel warna putih di sembarang tempat dalam gambar

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = image_2.copy()

# Ukuran kotak
box_width = 130  # lebar kotak
box_height = 50  # tinggi kotak

# Lokasi acak untuk kotak (koordinat kiri atas kotak)
start_x = 100
start_y = 90

# Menggambar kotak menggunakan kumpulan pixel putih
for y in range(start_y, start_y + box_height):
    for x in range(start_x, start_x + box_width):
        image_3[y, x] = [255, 255, 255]

# Gabungkan gambar asli dengan gambar yang memiliki kotak putih
final_frame = cv.hconcat((image_2, image_3))
cv2_imshow(final_frame)
```

```python
photos = ["foto1.jpg", "foto2.jpg", "foto3.jpg"]

# Pertama, perulangan untuk memproses gambar menggunakan OpenCV
for photo in photos:
    foto_ktm = io.imread(photo)  # Membaca foto KTM
    foto_ktm = cv.resize(foto_ktm, (0, 0), fx=0.5, fy=0.5)  # Resize foto KTM menjadi setengah ukuran
    foto_ktm_rgb = cv.cvtColor(foto_ktm, cv.COLOR_BGR2RGB)  # Konversi warna menjadi RGB

    # Menampilkan menggunakan OpenCV di Google Colab
    cv2_imshow(foto_ktm_rgb)  # Menampilkan gambar dengan format RGB
    print('OpenCV View')

# Kedua, perulangan untuk menampilkan gambar menggunakan Matplotlib
for photo in photos:
    foto_ktm = io.imread(photo)  # Membaca foto KTM
    foto_ktm = cv.resize(foto_ktm, (0, 0), fx=0.5, fy=0.5)  # Resize foto KTM menjadi setengah ukuran

    # Menampilkan gambar menggunakan Matplotlib
    plt.imshow(foto_ktm)
    plt.axis('off')  # Menghilangkan sumbu
    plt.show()  # Menampilkan gambar

    print('Matplotlib View')
    print('\n')  # Tambahkan baris baru setelah menampilkan foto
```
