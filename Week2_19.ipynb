4. Membuat bentuk Geometri 2D dari OpenCV. Diawali dengan pembuatan black image

2. Tampilkan image dalam channel Red-Blue dan Green-Blue saja!


```python
# Pisahkan channel warna dari gambar
B, G, R = cv.split(img)

# Saluran Red-Blue saja (Green = 0)
RB = cv.merge([B, np.zeros_like(G), R])

# Saluran Green-Blue saja (Red = 0)
GB = cv.merge([B, G, np.zeros_like(R)])

# Tampilkan kedua gambar
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# Tampilkan Red-Blue
ax[0].imshow(cv.cvtColor(RB, cv.COLOR_BGR2RGB))
ax[0].set_title('Red-Blue Channel')

# Tampilkan Green-Blue
ax[1].imshow(cv.cvtColor(GB, cv.COLOR_BGR2RGB))
ax[1].set_title('Green-Blue Channel')

plt.show()
```

5. Buat 5 kotak berbagai ukuran dan warna yang berbeda dalam satu image.
disarankan menggunakan bilangan acak/random!


```python
# Buat gambar kosong dengan ukuran 500x500 piksel
img = np.zeros((500, 500, 3), dtype=np.uint8)

# Tentukan jumlah kotak
num_boxes = 5

# Buat kotak dengan ukuran dan warna acak
for _ in range(num_boxes):
    # Warna acak dalam format BGR
    color = np.random.randint(0, 256, size=3).tolist()

    # Ukuran dan posisi acak
    x1 = np.random.randint(0, 400)
    y1 = np.random.randint(0, 400)
    x2 = x1 + np.random.randint(20, 100)
    y2 = y1 + np.random.randint(20, 100)

    # Gambarkan kotak pada gambar
    cv.rectangle(img, (x1, y1), (x2, y2), color, -1)  # -1 untuk mengisi kotak dengan warna

# Tampilkan gambar dengan kotak-kotak
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title('Image with Random Boxes')
plt.show()
```

7. Buat rectangle dan circle pada bagian wajah dari image foto anda saat
beraktifitas (bukan pasfoto).

```python
# Baca gambar
img = cv.imread('/content/drive/MyDrive/pcvk/prf.jpeg')

# Tambahkan rectangle (kotak)
top_left = (230, 200)  # Titik sudut kiri atas
bottom_right = (450, 280)  # Titik sudut kanan bawah
color_rectangle = (0, 255, 0)  # Warna kotak dalam format BGR (hijau)
thickness_rectangle = 2  # Ketebalan garis kotak
cv.rectangle(img, top_left, bottom_right, color_rectangle, thickness_rectangle)

# Tambahkan circle (lingkaran)
center = (350, 330)  # Titik tengah lingkaran
radius = 30  # Jari-jari lingkaran
color_circle = (255, 0, 0)  # Warna lingkaran dalam format BGR (merah)
thickness_circle = 2  # Ketebalan garis lingkaran, gunakan -1 untuk mengisi lingkaran
cv.circle(img, center, radius, color_circle, thickness_circle)

# Tampilkan gambar dengan rectangle dan circle
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.title('Image with Rectangle and Circle')
plt.axis('off')  # Nonaktifkan axis untuk tampilan yang lebih bersih
plt.show()
```

8. Buat rectangle pada bagian sudut bawah kiri channel B pada color space RGB

```python
# Baca gambar
img = cv.imread('/content/drive/MyDrive/pcvk/prf.jpeg')

# Pisahkan saluran BGR
B, G, R = cv.split(img)

# Tentukan ukuran gambar
height, width = B.shape

# Tentukan ukuran dan posisi kotak (rectangle) di sudut bawah kiri channel B
rect_width = 100
rect_height = 100
top_left = (0, height - rect_height)
bottom_right = (rect_width, height)

# Tambahkan rectangle pada channel Blue
cv.rectangle(B, top_left, bottom_right, (255), -1)  # Warna putih (255) dan ketebalan -1 untuk mengisi

# Gabungkan kembali saluran BGR
img_with_rectangle = cv.merge([B, G, R])

# Tampilkan gambar dengan kotak pada channel Blue
plt.imshow(cv.cvtColor(img_with_rectangle, cv.COLOR_BGR2RGB))
plt.title('Image with Rectangle in Blue Channel')
plt.axis('off')
plt.show()
```

9. Lengkapi tulisan nama file pada file citra dari soal no.8. gunakan font, ukuran
font, dan warna font yang sesuai keinginan anda.


```python
# Tambahkan teks ke gambar
text = 'Image with Rectangle'
font = cv.FONT_HERSHEY_SIMPLEX  # Jenis font
font_scale = 1  # Ukuran font
font_color = (0, 255, 0)  # Warna font (hijau dalam format BGR)
font_thickness = 2  # Ketebalan font

# Tentukan posisi teks
text_size, _ = cv.getTextSize(text, font, font_scale, font_thickness)
text_x = 100
text_y = height - 10

# Tambahkan teks ke gambar
cv.putText(img_with_rectangle, text, (text_x, text_y), font, font_scale, font_color, font_thickness)

# Tampilkan gambar dengan kotak dan teks
plt.imshow(cv.cvtColor(img_with_rectangle, cv.COLOR_BGR2RGB))
plt.title('Image with Rectangle and Text')
plt.axis('off')
plt.show()
```

## **Tugas Kelompok**


```python
# Daftar nama file gambar
image_paths = [
    '/content/drive/MyDrive/pcvk/ktm/foto1.jpg',
    '/content/drive/MyDrive/pcvk/ktm/foto2.jpg',
    '/content/drive/MyDrive/pcvk/ktm/foto3.jpg'
]

# Baca dan resize gambar ke dalam array
images = [cv.resize(cv.imread(image_path), (430, 300)) for image_path in image_paths]

# Warna dan koordinat lingkaran yang sama untuk setiap gambar
circle_data = [(100, 100, 20, (255, 0, 0)), (230, 200, 20, (0, 255, 0))]

# Tambahkan lingkaran yang sama pada setiap gambar
for img in images:
    for (x, y, radius, color) in circle_data:
        cv.circle(img, (x, y), radius, color, 3)

# Buat figure dengan 3 subplots
fig, ax = plt.subplots(1, 3, figsize=(15, 5))

# Tampilkan setiap gambar dengan lingkaran di subplot
for i, img in enumerate(images):
    ax[i].imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
    ax[i].set_title(f'Foto {i+1}')
    ax[i].axis('off')  # Nonaktifkan axis

# Tampilkan keseluruhan plot
plt.tight_layout()
plt.show()
```
